```python
import matplotlib.pyplot as plt
import numpy as np

# Данные: [Название, Количество токенов, Цвет]
data = [
    ("LLaMA 3",       15_000_000_000_000, "#6C5CE7"),
    ("GPT-4",          1_800_000_000_000, "#00B894"),
    ("Человек",              1_000_000_000, "#FD79A8"),
    ("Собака",                  50_000_000, "#FDCB6E"),
    ("Кошка",                   30_000_000, "#E17055"),
]

names  = [d[0] for d in data]
tokens = [d[1] for d in data]
colors = [d[2] for d in data]

# Размер пузырька пропорционален логарифму токенов (для наглядности)
sizes = [np.log10(t) * 800 for t in tokens]

fig, ax = plt.subplots(figsize=(10, 7))

# Строим пузырьковую диаграмму
scatter = ax.scatter(
    np.arange(len(names)),          # X-координаты
    np.zeros(len(names)),           # Y-координаты (все на одной линии)
    s=sizes,
    c=colors,
    alpha=0.7,
    edgecolors="black",
    linewidth=1.5,
    zorder=3
)

# Подписи под пузырьками
for i, (name, tok) in enumerate(zip(names, tokens)):
    # Форматируем число
    if tok >= 1e12:
        label = f"{tok/1e12:.1f}T"
    elif tok >= 1e9:
        label = f"{tok/1e9:.0f}B"
    elif tok >= 1e6:
        label = f"{tok/1e6:.0f}M"
    else:
        label = f"{tok:,}"

    ax.text(i, -0.02, name, ha='center', va='top', fontsize=12, fontweight='bold')
    ax.text(i, -0.06, label, ha='center', va='top', fontsize=10, color='gray')

# Настройки осей
ax.set_xlim(-0.5, len(names) - 0.5)
ax.set_ylim(-0.1, 0.1)
ax.axis('off')

ax.set_title(
    "Пузырьковая диаграмма: объём языковых токенов\n"
    "(логарифмическая шкала размеров)",
    fontsize=16, pad=20
)

# Легенда-подсказка
legend_elements = [
    plt.Line2D([0], [0], marker='o', color='w',
               markerfacecolor='#6C5CE7', markersize=15, label='LLaMA 3 (~15T)'),
    plt.Line2D([0], [0], marker='o', color='w',
               markerfacecolor='#00B894', markersize=12, label='GPT-4 (~1.8T)'),
    plt.Line2D([0], [0], marker='o', color='w',
               markerfacecolor='#FD79A8', markersize=10, label='Человек (~1B)'),
    plt.Line2D([0], [0], marker='o', color='w',
               markerfacecolor='#FDCB6E', markersize=8,  label='Собака (~50M)'),
    plt.Line2D([0], [0], marker='o', color='w',
               markerfacecolor='#E17055', markersize=7,  label='Кошка (~30M)'),
]
ax.legend(handles=legend_elements, loc='upper right', fontsize=10, framealpha=0.9)

plt.tight_layout()
# plt.savefig("token_bubble_chart.png", dpi=200)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Данные
# tokens: условный объем знаний
# vision: точность зрения (%)
# power: энергопотребление в Ваттах (Watt)
data = {
    'GPT-4': {'tokens': 13000, 'vision': 95, 'power': 5000, 'color': 'royalblue'},
    'Llama 3': {'tokens': 8000, 'vision': 85, 'power': 1200, 'color': 'skyblue'},
    'Claude 3': {'tokens': 11000, 'vision': 92, 'power': 4500, 'color': 'purple'},
    'Человек': {'tokens': 15000, 'vision': 100, 'power': 20, 'color': 'orange'},
    'Собака': {'tokens': 4000, 'vision': 75, 'power': 5, 'color': 'brown'},
    'Кошка': {'tokens': 3000, 'vision': 80, 'power': 3, 'color': 'gray'}
}

vision = [v['vision'] for v in data.values()]
tokens = [v['tokens'] for v in data.values()]
powers = [v['power'] for v in data.values()]
colors = [v['color'] for v in data.values()]
labels = list(data.keys())

# Масштабируем размеры пузырьков для наглядности (логическое разделение)
sizes = [p * 0.4 if p > 100 else p * 15 for p in powers]

plt.figure(figsize=(12, 8))

# Рисуем пузырьки
plt.scatter(vision, tokens, s=sizes, c=colors, alpha=0.6, edgecolors="black")

# Подписываем каждый объект
for i, name in enumerate(labels):
    plt.annotate(f"{name}\n({powers[i]}W)", 
                 (vision[i], tokens[i]),
                 textcoords="offset points", 
                 xytext=(0,15), 
                 ha='center', 
                 fontsize=10, 
                 fontweight='bold')

plt.title('Эффективность: Знания vs Зрение\n(Размер пузырька = Энергопотребление в Ваттах)', fontsize=15)
plt.xlabel('Точность распознавания изображений (%)', fontsize=12)
plt.ylabel('Условное количество токенов (Сложность)', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Данные в миллиардах (Billions)
# Для LLM: параметры + контекст
# Для существ: синапсы + визуальный процессинг (эквивалент)
data = {
    'Кошка': 0.8,
    'Собака': 2.2,
    'Llama 3 (70B)': 70,
    'GPT-4 (Estimation)': 1700,
    'Claude 3.5 (Est.)': 2000,
    'Человек': 150000  # 150 триллионов синапсов = 150,000 миллиардов
}

# Сортировка по возрастанию для наглядности
sorted_data = dict(sorted(data.items(), key=lambda item: item[1]))

names = list(sorted_data.keys())
values = list(sorted_data.values())

plt.figure(figsize=(12, 8))

# Используем логарифмическую шкалу, иначе Человек "съест" всех остальных
# и LLM превратятся в невидимые точки на фоне 150 триллионов.
plt.yscale('log') 

# Создаем пузырьковую диаграмму (вдоль одной линии для сравнения объемов)
colors = ['#8c564b', '#7f7f7f', '#1f77b4', '#9467bd', '#17becf', '#ff7f0e']
scatter = plt.scatter(names, values, s=[v*0.05 if v > 1000 else 500 for v in values], 
            c=colors, alpha=0.7, edgecolors="black")

# Добавляем подписи точных значений
for i, v in enumerate(values):
    label = f"{v:,} млрд" if v < 1000 else f"{v/1000:,.1f} трлн"
    plt.annotate(label, (names[i], values[i]), textcoords="offset points", 
                 xytext=(0,10), ha='center', fontweight='bold')

plt.title('Сравнение когнитивной емкости (в эквиваленте токенов/параметров)', fontsize=14)
plt.ylabel('Количество (Логарифмическая шкала)', fontsize=12)
plt.grid(True, which="both", ls="-.", alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Единая валюта: суммарные токены (Параметры + Контекст + Визуальный поток)
# Значения в миллиардах (1,000 = 1 триллион)
data = {
    'Кошка': 35,             # Синапсы + мощное зрение
    'Собака': 120,           # Больше нейронов в коре, чем у кошки
    'DeepSeek-V3': 671,      # Общее кол-во параметров (MoE)
    'GPT-4o': 1800,          # Оценочные 1.8 трлн параметров
    'Gemini 1.5 Pro': 2500,  # Огромный контекст + параметры
    'Человек': 150000        # ~150 триллионов синапсов (король токенов)
}

names = list(data.keys())
values = list(data.values())
colors = ['#7f7f7f', '#8c564b', '#d62728', '#2ca02c', '#1f77b4', '#ff7f0e']

# Создаем график
fig, ax = plt.subplots(figsize=(12, 8))

# Рисуем пузырьки (размер зависит от значения, но скейлим для красоты)
sizes = [np.sqrt(v) * 200 for v in values] 
scatter = ax.scatter(names, values, s=sizes, c=colors, alpha=0.6, edgecolors='black', linewidth=2)

# Устанавливаем логарифмическую шкалу, чтобы видеть всех одновременно
ax.set_yscale('log')

# Добавляем подписи данных
for i, name in enumerate(names):
    if values[i] >= 1000:
        label = f"{values[i]/1000:.1f} трлн"
    else:
        label = f"{values[i]} млрд"
    
    ax.annotate(label, (names[i], values[i]), 
                textcoords="offset points", 
                xytext=(0, 20), 
                ha='center', 
                fontsize=11, 
                fontweight='bold')

# Оформление
ax.set_title('Сравнение систем в "единой валюте" (Суммарные токены Параметры + Контекст + Визуальный поток)', fontsize=16, pad=30)
ax.set_ylabel('Количество токенов (Log Scale)', fontsize=12)
ax.grid(True, which="both", ls="--", alpha=0.4)

# Немного расширим границы, чтобы пузырьки не обрезались
ax.set_ylim(min(values)*0.5, max(values)*5)

plt.tight_layout()
plt.show()
```

# Критерии оценки и математическая модель «Единой валюты токенов»

Для сравнения биологических систем (от червя до человека) и искусственного интеллекта (LLM и роботов) на одной диаграмме все ресурсы были приведены к **единой валюте — эквивалентным токенам ($T_{total}$)**.

Итоговая емкость системы рассчитывается как сумма четырех компонент:
$$T_{total} = T_{genetic/hard} + T_{cognitive} + T_{vision} + T_{motor}$$

---

### 1. Предустановленный софт: Генетическая память ($T_{genetic}$) vs Цифровое хранилище ($T_{hard}$)

Компонента оценивает базовые системные инструкции, необходимые для функционирования «платформы».

*   **Биология ($T_{genetic}$):** Оценивается объем смысловой (кодирующей) информации в ДНК.
    *   *Математика:* 1 пара нуклеотидных оснований (bp) ДНК кодирует 2 бита информации (A, T, G, C). Информационный объем генома в байтах: $B = \frac{\text{Кол-во пар оснований} \times 2 \text{ бита}}{8 \text{ бит}}$.
    *   Принято, что 1 текстовый токен в среднем равен 4 байтам. Следовательно, формула перевода ДНК в токены:
    $$T_{genetic} = \frac{\text{Кол-во пар оснований}}{16}$$
    *   *Пример:* Геном человека (~3.2 млрд пар оснований) за вычетом некодирующих областей дает весомый вклад в архитектурную память системы. У простейших организмов (Амёба, Тихоходка) ДНК является главным источником токенов, так как содержит миллионы лет эволюционного «кода выживания» (до 15–50 млрд пар оснований).
*   **Искусственный Интеллект ($T_{hard}$):** Сумма весов модели на диске, объема операционной системы и низкоуровневых драйверов управления, переведенная в токены по коэффициенту $1 \text{ токен} = 4 \text{ байта}$.

---

### 2. Когнитивная емкость и контекст ($T_{cognitive}$)

Оценивает динамическую память, способность к обучению и удержанию логических связей.

*   **Биология:** Принята фундаментальная научная гипотеза: **1 синапс (связь между нейронами) $\approx$ 1 параметр искусственной нейросети**.
    *   *Математика:* Мозг человека содержит $\sim 86 \times 10^9$ нейронов. Каждый нейрон образует в среднем $10^3 - 10^4$ синапсов. Общее число синапсов у человека:
    $$\text{Синапсы} \approx 100 - 150 \text{ триллионов} (1.5 \times 10^{14})$$
    *   В модели 1 параметр/синапс приравнивается к 1 когнитивному токену базы знаний.
*   **Искусственный Интеллект:** Сумма активных параметров модели (для MoE-моделей, таких как DeepSeek-V3, берется общий объем в 671 млрд параметров) плюс размер контекстного окна (VRAM), умноженный на коэффициент плотности внимания.
    *   *Пример:* Gemini 1.5 Pro выделяется на графике за счет аномально большого контекстного окна в $2 \times 10^6$ токенов, работающего параллельно с базовыми параметрами.

---

### 3. Визуальный контекст и распознавание ($T_{vision}$)

Оценивает плотность данных, обрабатываемых зрительным процессором системы «здесь и сейчас».

*   **Биология:** Глаз человека передает по зрительному нерву около $10^7$ бит в секунду. Зрительная кора занимает около 15% всей коры головного мозга ($\sim 14 \times 10^9$ синапсов), непрерывно работая как мультимодальная сеть распознавания образов с частотой дискретизации $\sim 30-60$ Гц.
*   **Искусственный Интеллект:** Оценивается через механизм ViT (Vision Transformer). Одно изображение разрешением $512 \times 512$ пикселей разбивается на патчи и превращается примерно в $1024 - 2048$ визуальных токенов. При обработке видеопотока (мультимодальность в реальном времени) емкость контекста растет линейно каждую секунду.

---

### 4. Двигательные токены и моторика ($T_{motor}$)

Оценивает сложность удержания баланса в 3D-пространстве, проприоцепцию (кинестетическое считывание тела) и управление актуаторами/мышцами.

*   **Биология:** Моторика требует колоссальных вычислительных мощностей. У млекопитающих **мозжечок** (отвечающий за координацию и микромоторику) содержит до **80% всех нейронов мозга** (у человека — около 69 млрд из 86 млрд нейронов).
    *   *Математика:* Емкость мозжечка пересчитана через его синаптическую плотность:
    $$T_{motor\_bio} = \text{Кол-во синапсов мозжечка} \approx 10^{14} \text{ токенов}$$
    *   *Пример (Комар):* Насекомые обладают феноменальной частотой работы моторики. Комар совершает до 1000 взмахов крыльев в секунду, корректируя полет по тепловизору и гироскопическим органам (жужжальцам), что требует огромного «предустановленного» двигательного контекста ($180 \text{ млрд токенов}$).
*   **Искусственный Интеллект (Робототехника):** Оценивается через объем LBM (Large Behavior Models) — нейросетей, обучаемых управлению физическими объектами.
    *   *Пример:* Робот Atlas от Boston Dynamics использует современные нейросетевые контроллеры (модели порядка $4.5 \times 10^8$ параметров) в сочетании с низкоуровневыми процессорами обработки данных с лидаров и гироскопов, что дает суммарно около $50 \text{ млрд токенов}$.

### Резюме пропорций: Почему робот равен червю?

Натуральный масштаб диаграммы наглядно доказывает: **современная робототехника ($T_{Atlas} = 50 \text{ млрд}$) находится на уровне простейших микроорганизмов ($T_{Тихоходка} = 35 \text{ млрд}$)**.

Причина кроется в эволюционном парадоксе: чтобы заставить железное тело просто держать баланс и переставлять две ноги в меняющейся среде, инженерам приходится задействовать мощности, эквивалентные всей нервной системе и ДНК-инструкциям крошечного круглого червя (C. Elegans) или тихоходки. Природа за миллиарды лет упаковала сложнейшие алгоритмы выживания и управления физической материей гораздо плотнее, чем мы пока способны воспроизвести на кремниевых GPU.
