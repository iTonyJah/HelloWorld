```python
import matplotlib.pyplot as plt
import numpy as np

# Данные: [Название, Количество токенов, Цвет]
data = [
    ("LLaMA 3",       15_000_000_000_000, "#6C5CE7"),
    ("GPT-4",          1_800_000_000_000, "#00B894"),
    ("Человек",              1_000_000_000, "#FD79A8"),
    ("Собака",                  50_000_000, "#FDCB6E"),
    ("Кошка",                   30_000_000, "#E17055"),
]

names  = [d[0] for d in data]
tokens = [d[1] for d in data]
colors = [d[2] for d in data]

# Размер пузырька пропорционален логарифму токенов (для наглядности)
sizes = [np.log10(t) * 800 for t in tokens]

fig, ax = plt.subplots(figsize=(10, 7))

# Строим пузырьковую диаграмму
scatter = ax.scatter(
    np.arange(len(names)),          # X-координаты
    np.zeros(len(names)),           # Y-координаты (все на одной линии)
    s=sizes,
    c=colors,
    alpha=0.7,
    edgecolors="black",
    linewidth=1.5,
    zorder=3
)

# Подписи под пузырьками
for i, (name, tok) in enumerate(zip(names, tokens)):
    # Форматируем число
    if tok >= 1e12:
        label = f"{tok/1e12:.1f}T"
    elif tok >= 1e9:
        label = f"{tok/1e9:.0f}B"
    elif tok >= 1e6:
        label = f"{tok/1e6:.0f}M"
    else:
        label = f"{tok:,}"

    ax.text(i, -0.02, name, ha='center', va='top', fontsize=12, fontweight='bold')
    ax.text(i, -0.06, label, ha='center', va='top', fontsize=10, color='gray')

# Настройки осей
ax.set_xlim(-0.5, len(names) - 0.5)
ax.set_ylim(-0.1, 0.1)
ax.axis('off')

ax.set_title(
    "🫧 Пузырьковая диаграмма: объём языковых токенов\n"
    "(логарифмическая шкала размеров)",
    fontsize=16, pad=20
)

# Легенда-подсказка
legend_elements = [
    plt.Line2D([0], [0], marker='o', color='w',
               markerfacecolor='#6C5CE7', markersize=15, label='LLaMA 3 (~15T)'),
    plt.Line2D([0], [0], marker='o', color='w',
               markerfacecolor='#00B894', markersize=12, label='GPT-4 (~1.8T)'),
    plt.Line2D([0], [0], marker='o', color='w',
               markerfacecolor='#FD79A8', markersize=10, label='Человек (~1B)'),
    plt.Line2D([0], [0], marker='o', color='w',
               markerfacecolor='#FDCB6E', markersize=8,  label='Собака (~50M)'),
    plt.Line2D([0], [0], marker='o', color='w',
               markerfacecolor='#E17055', markersize=7,  label='Кошка (~30M)'),
]
ax.legend(handles=legend_elements, loc='upper right', fontsize=10, framealpha=0.9)

plt.tight_layout()
plt.savefig("token_bubble_chart.png", dpi=200)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Единая валюта (млрд токенов): Интеллект + Зрение + Моторика
data = {
    'Человек': 250000,
    'GPT-5 (Est.)': 10000,
    'Gemini 1.5 Pro': 2500,
    'GPT-4o': 1800,
    'Собака': 500,
    'Кошка': 300,
    'Atlas (Robot)': 1.5
}

colors = {
    'Человек': '#ff7f0e',
    'GPT-5 (Est.)': '#444444',
    'Gemini 1.5 Pro': '#1f77b4',
    'GPT-4o': '#2ca02c',
    'Собака': '#8c564b',
    'Кошка': '#7f7f7f',
    'Atlas (Robot)': '#d62728'
}

# Сортировка для правильного наслоения (маленькие поверх больших)
sorted_names = sorted(data, key=data.get, reverse=True)

plt.figure(figsize=(10, 10))
ax = plt.gca()

# Коэффициент масштабирования площади для визуализации
# s в matplotlib — это площадь в пунктах^2
scale = 0.5 

for name in sorted_names:
    value = data[name]
    ax.scatter(0, 0, s=value * scale, c=colors[name], label=f"{name}: {value:,} млрд", 
               alpha=0.7, edgecolors='black', linewidth=1.5, zorder=-value)

# Настройка осей (делаем их невидимыми, оставляем только центр)
ax.set_aspect('equal')
ax.axis('off')

# Лимиты, чтобы самый большой круг (Человек) вписался
limit = 350 
plt.xlim(-limit, limit)
plt.ylim(-limit, limit)

plt.title('Сводная диаграмма: Площадь = Сумма токенов\n(Интеллект, Зрение, Моторика)', fontsize=14, pad=20)
plt.legend(title="Субъекты", loc='center left', bbox_to_anchor=(1, 0.5), frameon=True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Единая валюта (млрд токенов): Когнитивка + Моторика + Генетика + Железо
data = {
    'Человек': 320000,
    'GPT-5 (Est.)': 15000,
    'Gemini 1.5 Pro': 4000,
    'GPT-4o': 2500,
    'Собака': 1200,
    'Кошка': 900,
    'Atlas (Robot)': 50
}

colors = {
    'Человек': '#ff7f0e',
    'GPT-5 (Est.)': '#444444',
    'Gemini 1.5 Pro': '#1f77b4',
    'GPT-4o': '#2ca02c',
    'Собака': '#8c564b',
    'Кошка': '#7f7f7f',
    'Atlas (Robot)': '#d62728'
}

# Сортировка от большего к меньшему
sorted_names = sorted(data, key=data.get, reverse=True)

fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# Масштабирование для 3D
# В matplotlib s — это 2D-площадь проекции. 
# Чтобы ОБЪЕМ сферы был пропорционален V, радиус R должен быть ~ V^(1/3).
# Тогда s (площадь круга на экране) должна быть пропорциональна R^2, т.е. V^(2/3).
for name in sorted_names:
    tokens = data[name]
    # Вычисляем размер маркера так, чтобы объем сферы визуально соответствовал токенам
    visual_size = (tokens**(2/3)) * 50 
    
    ax.scatter([0], [0], [0], s=visual_size, c=colors[name], 
               label=f"{name}: {tokens:,} млрд", alpha=0.6, 
               edgecolors='white', linewidth=0.5)

# Убираем оси и сетку для чистоты
ax.set_axis_off()

# Настройка ракурса
ax.view_init(elev=20, azim=45)

plt.title('3D Объем Интеллекта и Биологии\n(Объем сферы ∝ Количеству токенов)', fontsize=15, pad=0)
plt.legend(title="Объекты (млрд токенов)", loc='center left', bbox_to_anchor=(0.85, 0.5))

plt.tight_layout()
plt.show()
```
